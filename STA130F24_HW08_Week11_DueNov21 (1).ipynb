1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

**Accuracy:**

Accuracy measures the proportion of the correct predictions ( true positives and true negatives across the entire datasets), basically the proportion of total prediction that was correct. 

For example: Classifying products as 'defective' / 'non-defective' for quality control in manufacturing. 

Accuracy provide a holistic view of model performance in balanced dataset.

Another example: a medical test for a common cold that identifies both positive and negative cases ( for example, a covid test)

Accuracy is important as indicates the overall performance of a system. 

**Sensitivity:**

 Sensitivity measures the proportion of actual positives that are correclty identified ( the true positive rate)

For example: detecting cancer using a screening test. In medical applications, it is crucial to minimize false negatives (failing to identify someone who is sick) is very dangerous, as it can result in serious health consequences. 

**Specificity:**

Specificity  measure the actual true negative tht correctly identified. Specificity tells how well the model avoids false positives. 

For example, for a spam email fiter, specificity is important when minimizing false positives; false positive (incorrectly classifying a legitimate email as spam) are more disruptive than false negatives (letting some spam through). 

**Precision:**

Measures the proportion of positive predictions that are actually correct. Precision tells how reliable the positive predictions are.

For example, A fraud detection system for credit car transactions. 

In fraud detection, it’s important that when the system flags a transaction as fraudulent, it is very likely to be correct, as wrongly blocking legitimate transactions could frustrate customers and harm the business.

````python
import pandas as pd

# Load data
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Remove rows with NaN values (if required)
ab_reduced_noNaN = ab.dropna()

# Calculate the number of rows for the training set (80% of the data)
train_size = int(0.8 * len(ab_reduced_noNaN))  # 80% of the data

# Split the data: sample 80% for training, the remaining 20% for testing
ab_reduced_noNaN_train = ab_reduced_noNaN.sample(n=train_size, random_state=42)
ab_reduced_noNaN_test = ab_reduced_noNaN.drop(ab_reduced_noNaN_train.index)

# Print the number of observations in each dataset
print(f"Number of observations in the training dataset: {len(ab_reduced_noNaN_train)}")
print(f"Number of observations in the test dataset: {len(ab_reduced_noNaN_test)}")
````

# the number of observation in training data is 248, and the number of the observation in the test data is 62.

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Assuming you have already cleaned and reduced the dataset to 'ab_reduced_noNaN'

# Create the target variable `y` for hardcover or paperback (1 if hardcover, 0 if paperback)
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']

# Create the feature variable `X` with the List Price
X = ab_reduced_noNaN[['List Price']]

# Initialize the classifier with max_depth of 2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Train the classifier
clf.fit(X, y)

# Use tree.plot_tree to visualize the decision tree
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True, rounded=True)
plt.show()

# Optionally, print the decision tree's structure and how it makes predictions
print("Tree Structure:")
print(clf)

````

# Explanation

Target variable (y): We use pd.get_dummies() to convert the categorical "Hard_or_Paper" variable into a binary format, where 'H' represents the hardcover class (1) and the other class (paperback) is 0.

Feature variable (X): We select only the List Price variable as the predictor for the decision tree.

Classifier (clf): We use DecisionTreeClassifier() with a max_depth of 2, which restricts the tree to a maximum depth of 2 levels.

### 6. Use previously created *ab_reduced_noNaN_test* to create confusion matrices for *clf* and *clf2*. Report the sensitivity, specificity and accuracy for each of the models<br>

````python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score

# Assuming 'clf' and 'clf2' have already been trained and 'ab_reduced_noNaN_test' is available

# Create the target variable `y_test` for hardcover or paperback (1 if hardcover, 0 if paperback)
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

# Create the feature variable `X_test` for testing
X_test = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Get predictions for both models
y_pred_clf = clf.predict(ab_reduced_noNaN_test[['List Price']])
y_pred_clf2 = clf2.predict(X_test)

# Compute confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Display the confusion matrices without decimals
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=['Paperback', 'Hardcover'])
disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=['Paperback', 'Hardcover'])

# Plot the confusion matrices with no decimals in the values
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
disp_clf.plot(ax=axes[0], cmap='Blues', values_format='d')  # 'd' for integer format
disp_clf2.plot(ax=axes[1], cmap='Blues', values_format='d')  # 'd' for integer format
plt.show()

# Calculate metrics for clf
tn_clf, fp_clf, fn_clf, tp_clf = cm_clf.ravel()
sensitivity_clf = tp_clf / (tp_clf + fn_clf)  # Sensitivity: True Positive Rate
specificity_clf = tn_clf / (tn_clf + fp_clf)  # Specificity: True Negative Rate
accuracy_clf = accuracy_score(y_test, y_pred_clf)  # Accuracy

# Calculate metrics for clf2
tn_clf2, fp_clf2, fn_clf2, tp_clf2 = cm_clf2.ravel()
sensitivity_clf2 = tp_clf2 / (tp_clf2 + fn_clf2)  # Sensitivity: True Positive Rate
specificity_clf2 = tn_clf2 / (tn_clf2 + fp_clf2)  # Specificity: True Negative Rate
accuracy_clf2 = accuracy_score(y_test, y_pred_clf2)  # Accuracy

# Print the metrics with 3 decimal points
print(f"Metrics for clf:")
print(f"Sensitivity: {sensitivity_clf:.3f}")
print(f"Specificity: {specificity_clf:.3f}")
print(f"Accuracy: {accuracy_clf:.3f}")

print(f"\nMetrics for clf2:")
print(f"Sensitivity: {sensitivity_clf2:.3f}")
print(f"Specificity: {specificity_clf2:.3f}")
print(f"Accuracy: {accuracy_clf2:.3f}")
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions.
