````python
import numpy as np

# Sample data 
sample = np.array([12, 7, 9, 15, 11, 13, 10, 14, 8, 16])

# Number of bootstrap samples ( number of time that we are repeating bootstrapping)
n_iterations = 10000

# Create an array to store collections of means 
bootstrap_means = np.empty(n_iterations)

# Length of the sample
n = len(sample)

# Generate bootstrap samples and calculate the mean for each sample
for i in range(n_iterations):
    # Create a bootstrap sample by sampling with replacement from the original sample
    bootstrap_sample = np.random.choice(sample, size=n, replace=True)
    
    # Compute the mean of the bootstrap sample
    bootstrap_means[i] = np.mean(bootstrap_sample)

# Compute the 2.5th and 97.5th percentiles of the bootstrap means (for a 95% CI)
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)

# Display the 95% confidence interval
print(f"95% Confidence Interval for the Mean: [{lower_bound}, {upper_bound}]")
````

Explanation:

A confidence interval gives a range of plausible values for the true population parameter. If zero is within this range, it suggests the drug's effect could plausibly be zero, meaning we "fail to reject the null hypothesis" that there is no effect.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV file
df = pd.read_csv('health_scores.csv')

# Calculate the difference between FinalHealthScore and InitialHealthScore
df['Difference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Set seed for reproducibility
np.random.seed(42)

# Bootstrap function to sample the mean differences
def bootstrap_mean(data, n_bootstrap=10000):
    bootstrap_means = []
    for _ in range(n_bootstrap):
        sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_means.append(np.mean(sample))
    return bootstrap_means

# Perform bootstrapping on the 'Difference' column
boot_means = bootstrap_mean(df['Difference'].values)

# Plot the histogram of bootstrapped means
plt.figure(figsize=(10, 6))
plt.hist(boot_means, bins=30, color='skyblue', edgecolor='black')
plt.title('Bootstrap Distribution of the Mean Health Score Differences')
plt.xlabel('Mean Difference in Health Score (Final - Initial)')
plt.ylabel('Frequency')
plt.axvline(np.mean(boot_means), color='red', linestyle='dashed', linewidth=1)
plt.show()

# Confidence interval (95%)
confidence_interval = np.percentile(boot_means, [2.5, 97.5])

# Print the results
print(f"Mean difference in health scores from bootstrap: {np.mean(boot_means)}")
print(f"95% Confidence Interval: {confidence_interval}")
````

**According to bootstrape results, we are 95% confidence that the population parameter (mean difference ) lie within 0.9 and 5.6; 0 is excluded, therefore we can conclude to reject null hypothesis, and the vaccine has posistive effect on people's health.**
