**Q1**

Simple line regression is a statistical model use to predict the relationship between 2 continuous variables, the independent variable X and the dependent variable Y are assumed to be linear. 

we use X to predict Y

the slope coefficient is the rate of change of Y in terms of unit change of X.

intercept is the Y value when the X is zero (the baseline for our line)

Error term is the term that captures any variability in Y that is not explianed by X. It is normally assumed as a normal distribution with mean of 0 and standard deviation σ. 

Y=β0+β1X+ϵ

we assume the values will vary around the regression line according to the error term, therefore, when the error term is normally distributed, the obsered value Y will also be normally distributed around the regression line. 


````python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import uniform, norm

# Parameters for the model
n = 100              # Number of samples
beta0 = 3            # Intercept
beta1 = 1.2          # Slope
sigma = 1.0          # Standard deviation of errors

# Generate predictor values (X) and error terms (epsilon)
X = uniform.rvs(0, 10, size=n)      # X values from a uniform distribution [0, 10]
epsilon = norm.rvs(0, sigma, size=n) # Normally distributed errors

# Calculate outcome values (Y) based on the linear model
Y = beta0 + beta1 * X + epsilon

# Visualization
fig = go.Figure()
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Data points'))
fig.add_trace(go.Scatter(x=np.sort(X), y=beta0 + beta1 * np.sort(X), 
                         mode='lines', name='Theoretical Line',
                         line=dict(color='red')))
fig.update_layout(title="Simple Linear Model: Y = beta0 + beta1*X + epsilon",
                  xaxis_title="Predictor Variable (X)",
                  yaxis_title="Outcome Variable (Y)")
fig.show(renderer="png")
````

A **simulated** version of the "long wait times" dataset can be created by first creating `old_faithful_simulation = old_faithful[long_wait].copy()` and then assigning the **simulated** it values with `old_faithful_simulation['duration'] = 1.65 + 0*old_faithful_simulation.waiting + stats.norm(loc=0, scale=0.235).rvs(size=long_wait.sum())`

````python
import seaborn as sns
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load dataset
old_faithful = sns.load_dataset('geyser')

# Identify the long wait times (assuming 'waiting' >= 63 minutes corresponds to long waits)
long_wait = old_faithful['waiting'] >= 63

# Step 1: Bootstrapping to get the slope coefficients
n_bootstraps = 1000  # Number of bootstraps
bootstrapped_slope_coefficients = []

# Create bootstrap samples and fit the model
for _ in range(n_bootstraps):
    # Create a bootstrapped sample of long wait times
    bootstrap_sample = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    
    # Fit the simple linear regression model
    model = smf.ols('duration ~ waiting', data=bootstrap_sample)
    fitted_model = model.fit()
    
    # Collect the slope coefficient
    bootstrapped_slope_coefficients.append(fitted_model.params['waiting'])

bootstrapped_slope_coefficients = np.array(bootstrapped_slope_coefficients)

# Step 2: Visualize the bootstrapped sampling distribution of the slope coefficients
plt.figure(figsize=(10, 6))
plt.hist(bootstrapped_slope_coefficients, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(np.mean(bootstrapped_slope_coefficients), color='red', linestyle='dashed', linewidth=2, label="Mean of Bootstrap Coefficients")
plt.title("Bootstrapped Sampling Distribution of the Slope Coefficients")
plt.xlabel("Slope Coefficients (β1)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Step 3: Calculate the 95% Confidence Interval
ci_95 = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print(f"95% Bootstrapped Confidence Interval for the Slope Coefficients: {ci_95}")

# Step 4: Simulate the dataset under the null hypothesis (β0 = 1.65, β1 = 0)
n_simulations = 1000  # Number of simulations
simulated_slope_coefficients = []

# Simulate the data under the null hypothesis
for _ in range(n_simulations):
    # Create simulated version of the dataset
    old_faithful_simulation = old_faithful[long_wait].copy()
    old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation['waiting'] + stats.norm(loc=0, scale=0.37).rvs(size=long_wait.sum())
    
    # Fit the model on the simulated data
    model_sim = smf.ols('duration ~ waiting', data=old_faithful_simulation)
    fitted_model_sim = model_sim.fit()
    
    # Collect the slope coefficient from the simulated model
    simulated_slope_coefficients.append(fitted_model_sim.params['waiting'])

simulated_slope_coefficients = np.array(simulated_slope_coefficients)

# Step 5: Calculate the p-value for the simulated sampling distribution of the slope coefficient
observed_slope = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params['waiting']
p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print(f"Simulated p-value: {p_value}")

# Step 6: Check if 0 is within the 95% bootstrapped confidence interval
is_zero_in_ci = 0 >= ci_95[0] and 0 <= ci_95[1]
print(f"Is 0 within the 95% Bootstrapped Confidence Interval? {is_zero_in_ci}")

# Step 7: Compare the p-value from simulations to the p-value from the original model
original_model = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()
print(f"Original Model p-value: {original_model.pvalues['waiting']}")
````
